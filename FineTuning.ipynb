```python
From a csv file, use the tool to prepare the data into a JSONL file
```

```python
openai tools fine_tunes.prepare_data -f D:\Uni\Stuttgart Uni\Openai\Train_Model.csv
```

```python
After that, using the api, we create our fine tuned model, based in davinci due to its results on coding
```

```python
openai api fine_tunes.create -t D:\Uni\Stuttgart Uni\Openai\Train_Model_prepared.jsonl -m davinci
```

```python
And finally, we are able to use it as model with the following command to generate completions.
```

```python
openai api completions.create -m davinci:ft-hdm-university-stuttgart-2022-01-24-20-12-41 -p "Translate this prompt to SQL: what percent of users who signed up converted to paid?->"
```
